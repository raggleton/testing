```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax1 = plt.subplots()
time = [0, 1, 2]
pos = [1, 2, 4]
laser = [10, 30, 40]
ax1.plot(time, pos, 'mediumblue')
ax1.set_xlabel('Time / s')
# Make the y-axis label and tick labels match the line color.
ax1.set_ylabel('Displacement / nm', color='mediumblue')
for tl in ax1.get_yticklabels():
    tl.set_color('mediumblue')

ax2 = ax1.twinx()
ax2.plot(time, laser, 'crimson')
ax2.set_ylabel('Laser drive voltage / V', color='crimson')
for tl in ax2.get_yticklabels():
    tl.set_color('crimson')
```
